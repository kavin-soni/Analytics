# Principal Component Analysis

Principal component analysis is an algorithm for dimensionality reduction using feature extraction.
It combines input variables in a specific way, so that we can drop the least important variables and retain the most significant variables.

PCA can make remarkable improvements in the model performance when there are large number of variables.

read more about PCA here:
https://towardsdatascience.com/a-one-stop-shop-for-principal-component-analysis-5582fb7e0a9c

I will use logistic regression with raw data, standardized data, and PCA transformed data. Then I will compare the model performance.

```python
# Importing libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
# loading dataset from scikit
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
df = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
#(['DESCR', 'data', 'feature_names', 'target_names', 'target'])
```

## Data overview

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

**We can notice that there are 30 variables that can affect the likelyhood of causing cancer**


This makes it a perfect dataset for using PCA to reduce dimension and increase accuracy.

## Logistic Regression


```python
X = df
y = cancer['target']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression()
```

```python
log.fit(X_train,y_train)
```

```python
predict_initial = log.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report
```

```python
a = accuracy_score(y_test,predict_initial)
a
```

#### We get 93.5% accuracy with the raw data. This is not bad but let's use standard scaled data

## Standard Scaled Data 

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaled_df = scaler.fit(df)
scaled_data = scaler.transform(df)
```

```python
df_scaled = pd.DataFrame(scaled_data,columns=cancer['feature_names'])
df_scaled.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaled_data, y, test_size=0.30)
log_std = log.fit(X_train,y_train)
predict_2 = log_std.predict(X_test)
```

```python
b = accuracy_score(y_test,predict_2)
b
```

#### We notice that the accuracy increased slightly to 96% using stadardized data set . Not bad!

## PCA Trasformed Data

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA()
pca.fit(df_scaled)
```

```python
X_pca = pca.transform(df_scaled)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size=0.30)
log.fit(X_train,y_train)
predict_3 = log.predict(X_test)
```

```python
c = accuracy_score(y_test,predict_3)
c
```

#### The Accuracy increased to a whopping 98.8%. This is remarkable improvement.

## Model comparision 

```python
accuracy = []
accuracy.append(a)
accuracy.append(b)
accuracy.append(c)

methods = ['Direct data','Standardized data','Stardardized data with PCA']
```

```python
acc_df = pd.DataFrame(accuracy,methods)
acc_df
```
